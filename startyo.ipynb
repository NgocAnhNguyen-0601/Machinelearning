# Feature Selection:


Drop the Adminisatration feature due to the weak correlation with Profit (which the model want to predict)

```python
X = data[['Marketing Spend', 'R&D Spend']]
Y = data['Profit']
```

# Train and Test Split:


```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=40)
```

# Multiple Linear Regression Model:

```python
from sklearn.linear_model import LinearRegression
Linearreg = LinearRegression()
Linearreg.fit(X_train, Y_train)
```

# Predictions and Model Evaluation Based on Test Data:

```python
Y_pred_test = Linearreg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print("R2 Score: ", r2_score(Y_test, Y_pred_test))
print("Mean Absolute Error: ", mean_absolute_error(Y_test, Y_pred_test))
print("Mean Squared Error: ", mean_squared_error(Y_test, Y_pred_test))
print("Root Mean Squared Error: ", np.sqrt(mean_squared_error(Y_test, Y_pred_test)))
if hasattr(Linearreg, 'intercept_'):
    print("Intercept: ", Linearreg.intercept_)
else:
    print("The model does not have an intercept.")

if hasattr(Linearreg, 'coef_'):
    print("Coefficients: ", Linearreg.coef_)
else:
    print("The model does not have coefficients (e.g., if no features were used).")
```

```python
plt.figure(figsize=(8, 6))
plt.grid(False)
sns.scatterplot(x = Y_test, y = Y_pred_test, color = 'blue')
plt.xlabel('Actual Profit', fontsize=12, fontweight='bold')
plt.ylabel('Predicted Profit', fontsize=12, fontweight='bold')
plt.title('Actual vs Predicted Profit', fontsize=14, fontweight='bold')
```

# Predictions and Model Evaluation Based on Training Data:

```python
Y_pred_train = Linearreg.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(Y_train, Y_pred_train)
mae = mean_absolute_error(Y_train, Y_pred_train)
r2 = r2_score(Y_train, Y_pred_train)
print("Train Mean Squared Error: ", mse)
print("Train Mean Absolute Error: ", mae)
print("Train R2 Score: ", r2)
if hasattr(Linearreg, 'intercept_'):
    print("Intercept: ", Linearreg.intercept_)
else:
    print("The model does not have an intercept.")

if hasattr(Linearreg, 'coef_'):
    print("Coefficients: ", Linearreg.coef_)
else:
    print("The model does not have coefficients (e.g., if no features were used).")
```
